# Experiment 1
### Generate random samples from various probability distributions using Python

```python
# Import Required Libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
np.random.seed(42)
```

## Aim:
Generate random samples from:
1. Normal Distribution
2. Uniform Distribution
3. Binomial Distribution
4. Poisson Distribution

```python
# Generate Normal Distribution Samples

normal_data = np.random.normal(loc=50, scale=10, size=1000)

print("First 10 Normal Samples:")
print(normal_data[:10])
```

```python
# Plot Normal Distribution

plt.figure(figsize=(8,5))
sns.histplot(normal_data, bins=30, kde=True, color='green')
plt.title("Normal Distribution (Mean=50, Std=10)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Generate Uniform Distribution Samples

uniform_data = np.random.uniform(low=0, high=100, size=1000)

print("First 10 Uniform Samples:")
print(uniform_data[:10])
```

```python
# Plot Uniform Distribution

plt.figure(figsize=(8,5))
sns.histplot(uniform_data, bins=30, kde=True, color='blue')
plt.title("Uniform Distribution (0 to 100)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Generate Binomial Distribution Samples

binomial_data = np.random.binomial(n=10, p=0.5, size=1000)

print("First 10 Binomial Samples:")
print(binomial_data[:10])
```

```python
# Plot Binomial Distribution

plt.figure(figsize=(8,5))
sns.histplot(binomial_data, bins=11, kde=False, color='red')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Successes")
plt.ylabel("Frequency")
plt.show()
```

```python
# Generate Poisson Distribution Samples

poisson_data = np.random.poisson(lam=4, size=1000)

print("First 10 Poisson Samples:")
print(poisson_data[:10])
```

```python
# Plot Poisson Distribution

plt.figure(figsize=(8,5))
sns.histplot(poisson_data, bins=15, kde=False, color='yellow')
plt.title("Poisson Distribution (Lambda=4)")
plt.xlabel("Events")
plt.ylabel("Frequency")
plt.show()
```

```python
# Summary Statistics

datasets = {
    "Normal": normal_data,
    "Uniform": uniform_data,
    "Binomial": binomial_data,
    "Poisson": poisson_data
}

for name, data in datasets.items():
    print(f"{name} Distribution")
    print("Mean:", np.mean(data))
    print("Variance:", np.var(data))
    print("-"*30)
```

## Result Analysis

- Normal distribution gives bell-shaped curve.
- Uniform distribution gives flat spread.
- Binomial gives discrete bars.
- Poisson gives right-skewed discrete count distribution.
